```python
import numpy as np
import matplotlib.pyplot as plt
```

자기전에 2026년 수능 3문제만 풀어보기

```python
a = 3  #a를 바꿔가며 확인해보기...(아쉽긴 함, 그리고 답이 바로 암산이 되서 그렇기도)
f = lambda x: 3*x-2
g = lambda x: x**2-3*x+a
x = np.linspace(0,2,100)
plt.plot(x,f(x))
plt.plot(x,g(x))
```

```python
f = lambda x: (x+2)*(2*x**2-x-2)
h = 0.000001
(f(1+h)-f(1))/h
```

```python
#x가 0에서 2일때, a곡선 적분값-b곡선 적분값
f = lambda x: x**2+3
g = lambda x: -(1/5)*x**2+3
x = np.linspace(0,2,1000)
np.mean(f(x))*2 - np.mean(g(x))*2
```
